```python
G=10
U=0.0004
s=0.1
N=1e4
beta=1/5000.
r=0.
pi=1
tau=10
p0=1./N
assert r == 0, "Recombination not implemented"
```

```python
def del_mut_mat(U,G,pi=0,tau=1):
    '''del mut matrix'''
    dM = zeros((G,G))
    for j in range(G):
        v = stats.poisson([U + U*(tau - 1)*(i >= pi) for i in range(0,G-j)]).pmf(j)
        dM += diag(v, -j)
    dM[-1,:] += (1 - dM.sum(axis=0))
    for j in range(G):
        if allclose(dM[-1,j],1): dM[-1,j]=1
        elif allclose(dM[-1,j],0): dM[-1,j]=0
    assert allclose(1, dM.sum(axis=0)), "All columns must sum to one"
    return dM
dM = del_mut_mat(U,G,0,1)
assert all(del_mut_mat(10*U,G,0,1)==del_mut_mat(U,G,0,10))

def ben_mut_mat(beta,U,G,pi=0,tau=1):
    '''ben mut matrix'''
    bM = zeros((G,G))
    for j in range(G):
        v = stats.poisson([beta * (U + U*(tau - 1)*(i >= pi)) for i in range(j,G)]).pmf(j)
        bM += diag(v, j)
    bM[0,:] += (1 - bM.sum(axis=0))
    for j in range(G):
        if allclose(bM[-1,j],1): bM[-1,j]=1
        elif allclose(bM[-1,j],0): bM[-1,j]=0
    assert allclose(1, bM.sum(axis=0)), "All columns must sum to one"
    return bM
bM = ben_mut_mat(beta,U,G,0,1)
assert all(ben_mut_mat(beta,10*U,G,0,1)==ben_mut_mat(beta,U,G,0,10))

def sel_mat(s,G):
    '''selection matrix'''
    w = (1 - s)**arange(G)
    S = diag(w)
    assert (S * (ones((G,G)) - eye(G)) == 0).all(), "Non main diagonal elements must be zeros"
    return S
S = sel_mat(s,G)
```

```python
def evolve(t, p, condition, N, S, dM, bM, verbose=True):
    N = int(N)
    if verbose:
        W = [S.dot(p).sum()]
        P = [p.copy()]
    while condition(t,p):
        assert all(p >= 0), "frequencies vector must be non-negative (tick %d, a)"%t
        p[:] = S.dot(p)
        assert all(p >= 0), "frequencies vector must be non-negative (tick %d, b)"%t
        p[:] = dM.dot(p)
        assert all(p >= 0), "frequencies vector must be non-negative (tick %d, c)"%t
        p[:] = bM.dot(p)
        assert all(p >= 0), "frequencies vector must be non-negative (tick %d, d)"%t
        p[:] /= p.sum()
        assert allclose(p.sum(), 1), "frequencies vector must sum to one (tick %d, e), sum to %.3f" % (t, p.sum())
        
        if N > 0:
            p[:] = multinomial(N, p) / float(N)
        assert allclose(p.sum(), 1), "frequencies vector must sum to one (tick %d, f), sum to %.3f" % (t, p.sum())
        t += 1
        if verbose:
            W.append(S.dot(p).sum())
            P.append(p.copy())
    if verbose:
        return t,P,W
    else:
        return t,p
```

```python
def run(wait_ticks,N,s,U,beta,G,pi,tau,PLOT=True):
    color = rcParams['axes.color_cycle'][int(log10(wait_ticks))]
    pfix = exp(-U/s)

    ## initialize 
    dM = del_mut_mat(U,G,pi,tau)
    bM = ben_mut_mat(beta,U,G,pi,tau)
    zM = ben_mut_mat(1e-20,U,G,pi,tau)
    S = sel_mat(s,G)

    MSB_ticks = 1000#int(log(0.001) / log(1 - s)) * 2
    p = zeros(G)
    p[0] = 1.

    ## MSB
    t,P,W = evolve(0, p, lambda t,p: t < MSB_ticks, N, S, dM, bM)

    ## change environment
    p = P[-1]
    p[1:] = p[:-1]
    p[0] = 0

    # waiting time
    t,P_,W_ = evolve(t, p, lambda t,p: t < (wait_ticks+MSB_ticks), N, S, dM, zM)
    if PLOT: 
        axvline(x=t, color=color, ls='--', lw=3)
        m = average(range(G), weights=P_[-1])
        axhline(y=m, xmin=t, color=color, ls='--', lw=3)
    P.extend(P_)
    W.extend(W_)

    # appearance of beneficial mutation
    p[0] = p0
    p[1] -= p0

    ## fixation or extinction of mutant
    t,P_,W_ = evolve(t, p, lambda t,p: 0 < p[0] < pfix, N, S, dM, bM)
    P.extend(P_)
    W.extend(W_)

    M = [average(range(G), weights=x) for x in P]

    # plot
    if PLOT: 
        plot(range(t+3), M, color=color, label=str(wait_ticks))
        xlabel('t')
        ylabel('# mutations')
        d = {'$U/s$':U/s, '$1$':1, '$1+U/s$':1+U/s}
        if tau > 1:
            d.update({r'$\tau U/s$':tau*U/s, r'$1+\tau U/s$':1+tau*U/s})
        for k,y in d.items():
            axhline(y=y, color='k', ls='--', lw=0.5)
            text(4100, y, k)
        sns.despine()
    return t,P,W,M
```

## Results

First we will see the adaptation process in a non-mutator (NM) population.

The population starts mutation-free, proceeds to MSB, then an environmental change occurs which effectively adds a single deleterious mutation to all individuals. Then the population waits a set number of generations $T_w$ before appearance of the beneficial mutant, which then proceeds to deterministic fixation (without drift).

```python
pi,tau = 0,1
run(1,N,s,U,beta,G,pi,tau)
run(10,N,s,U,beta,G,pi,tau)
run(100,N,s,U,beta,G,pi,tau)
run(1000,N,s,U,beta,G,pi,tau)
lg = legend(title="$T_w$", loc='upper left')
lg.get_frame().set_facecolor('white')
ylim(0,1.5)
xlim(0,4000)
title("NM");
```

You can see that after the environmental change, the average number of mutations (y-axis) changes from $U/s$ to $1+U/s$, stays there until the appearance of the beneficial mutant at generation $T_w$ (vertical colored dashed line), and then the number of deleterious mutations slowly decreases back to $U/s$ as the mutant fixates.

Next, in a stress-induced mutator (SIM) populations with $\pi=1, \tau=10$:

```python
pi,tau = 1,10
run(1,N,s,U,beta,G,pi,tau)
run(10,N,s,U,beta,G,pi,tau)
run(100,N,s,U,beta,G,pi,tau)
run(1000,N,s,U,beta,G,pi,tau)
lg = legend(title="$T_w$", loc='upper left')
lg.get_frame().set_facecolor('white')
ylim(0,4.5)
xlim(0,4000)
title("SIM");
```

Here, after the environmental change the population evolves towards the MSB with $1++\tau U/s$ deleterious mutations, on average, because the environmental change causes it to hypermutate. 

However, the approach to this new MSB takes time, and if the beneficial mutant appears before the population reached the MSB (as it occurs for $T_w < 1000$ here) then the average number of deleterious mutations is still somewhere between $1+U/s$ and $1 + \tau U/s$.

The fixation probability $p_{fix}$ is positively correlated with the average number of mutations $m$, which is positively correlated with the hypermutation rate $\tau U$ and negatively correlated with the waiting time $T_w$, which in turn is negatively correlated with the hypermutation rate $\tau U$:
$$
p_{fix} \sim m \sim \tau, T_w \\
T_w \sim \frac{1}{\tau}
$$
and adaptation rate $\nu$ depends both on fixation probability and waiting time:
$$
\nu = \frac{p_{fix}}{T_w}
$$
so the total adaptation rate is correlated both with $\tau$ and with $\frac{1}{\tau}$, which suggest that there is an optimum $\tau$.

```python
def wait(N,s,U,beta,G,pi,tau,PLOT=True):
    ## initialize 
    dM = del_mut_mat(U,G,pi,tau)
    bM = ben_mut_mat(beta,U,G,pi,tau)
    zM = ben_mut_mat(1e-20,U,G,pi,tau)
    S = sel_mat(s,G)

    MSB_ticks = 1500#int(log(0.001) / log(1 - s)) * 2
    p = zeros(G)
    p[0] = 1.

    ## MSB
    t,P,W = evolve(0, p, lambda t,p: t < MSB_ticks, N, S, dM, bM)

    ## change environment
    p = P[-1]
    p[1:] = p[:-1]
    p[0] = 0

    # waiting time
    t,P_,W_ = evolve(t, p, lambda t,p: t < (2*MSB_ticks), N, S, dM, zM)
    P.extend(P_)
    W.extend(W_)

    M = [average(range(G), weights=x) for x in P]

    # plot
    if PLOT: 
        plot(range(t+2), M)
        xlabel('t')
        ylabel('# mutations')
        d = {'$U/s$':U/s, '$1$':1, '$1+U/s$':1+U/s}
        if tau > 1:
            d.update({r'$\tau U/s$':tau*U/s, r'$1+\tau U/s$':1+tau*U/s})
        for k,y in d.items():
            axhline(y=y, color='k', ls='--', lw=0.5)
            text(2500, y, k)
        ylim(0, 1+tau*U/s+0.5)
        sns.despine()
    return t,P,W,M
```

From Gordo and Dionisio (2005) (see [blog post](http://blog.yoavram.com/convergence-of-mean-fitness/)):
$$
m(t) = \frac{U}{s}(1-(1-s)^t)
$$
where $m(t)$ is the average number of mutations in the population at time $t$.

We will use for SIM this formula for the avg. number of mutation $t$ generations after the environmental change (given that it occured after the MSB was reached):
$$
m(t) = 1+\frac{U}{s} + (\tau-1)\frac{U}{s} (1 - (1-s)^t)
$$
The first term gives the number of mutations at the initial (NM) MSB, plus the additional mutation caused by the environmental change.
The second term reflects the increased mutational load caused by hypermutation in SIM with $\pi=1$ populations which increase the mutation rate $\tau$-fold. This mutational load is accumulated according to the formula given by Gordo and Dionisio above.

So in general we can estimate the mean fitness of SIM population $t$ generations after the environmental change using the above expression for $m(t)$:

$$
\bar{\omega}(t) = \sum_{k=0}^{\infty}{(1-s)^k e^{-m(t)} \frac{m(t)^k}{k!}} = \\
e^{-m(t)} \sum_{k=0}^{\infty}{\frac{((1-s)m(t))^k}{k!}} = \\
e^{-m(t)} e^{(1-s)m(t)} = e^{-m(t) + (1-s)m(t)} =  e^{-m(t) + m(t) - sm(t)} =  e^{-sm(t)} = \\
e^{-s-U - (\tau-1)U (1 - (1-s)^t)} \approx (1-s) \cdot e^{-U} \cdot e^{- (\tau-1)U (1 - (1-s)^t)}
$$

Of course when $\tau=1$ then $t$ is irelevant, because the population started from the NM MSB. 

When $t \to \infty$, $(1-s)^t \to 0$ and the mean fitness approaches $\bar{\omega}^* = (1-s) \cdot e^{-\tau U}$.

```python
U,s,tau = 0.003, 0.1, 10

def mean_fitness(t,s,U,tau):
    return (1-s) * exp(-U - (tau - 1) * U * (1 - (1 - s)**t))

plot(mean_fitness(arange(200), s, U, 1), label='NM')
plot(mean_fitness(arange(200), s, U, tau), label='CM')
axhline(y=exp(-U)*(1-s), ls=':', lw=3, color='gray')
axhline(y=exp(-tau*U)*(1-s), ls=':', lw=3, color='gray')
xlabel('time')
ylabel(r'$\bar{\omega}$')
legend(loc='center right')
axvline(x=50, color='k', ls='--', alpha=0.6)
sns.despine()
```

## Effect on fixation probability

Consider a beneficial mutant that appears $t$ generations after the environmental change (which occured after the population reached an MSB). 

The fixation probability can be defined by the following recursion. We ignore mutation, considering only selection and drift. 

$q(t,x)$ is the extinction probability of the mutant when in time $t$ the frequnecy of the mutant is $x$ in a population of size $N$, the fitness of the mutant is $\omega$, and the population mean fitness is $\bar{\omega}(t)$:

$$
q(t,k) = 
\sum_{i=0}^{N}{ \varphi_{N,i}\Big( \frac{k}{N} \cdot \frac{\omega}{\bar{\omega}(t,\frac{k}{N})} \Big) \cdot q\Big(t+1,i\Big) }
$$

where:
1. $q(t,0) = 1$
1. $\varphi_{N,k}(x)$ is the probability to for $k$ mutant offsprings from $N$ parents with mutant frequency $x$ (Binomial distribution):
$$
\varphi_{N,k}(x) = {N \choose k} x^k (1-x)^{N-k}
$$
1. $\bar{\omega}(t,x)$ is the population mean fitness at time $t$ with frequency of mutants $x$:
$$
\bar{\omega}(t,x) = x \cdot \omega + (1-x) \cdot (1-s) \cdot e^{-U} \cdot e^{- (\tau-1)U (1 - (1-s)^t)} \\
\omega = e^{-U}
$$
1. $\bar{\omega}^*(x) = \lim_{t \to \infty}{\bar{\omega}(t,x)} = x \cdot \omega + (1-x)(1-s) e^{-\tau U}$
1. $\lim_{t \to \infty}{q(t,x)} = q_{kim}(N,x,\frac{\omega}{\bar{\omega}^*(x)}-1)$ with $q_{kim}$ being the complement of _Kimura_'s fixation probability: $q_{kim}(N,x,s) = 1 - \frac{1-exp(-2Nsx)}{1-exp(-2Ns)}$.

Because the mutant is beneficial and $\bar{\omega}(t,x)$ decreases in $t$ (therefore $\frac{\omega}{\bar{\omega}(t,x)}$ increases in $t$) we expect $q(t,x)$ to decrease both in $x$ and in $t$.

```python
def q_msb(x,N,s,U,tau):
    assert 1 <= N
    assert 0 <= x <= 1
    assert 0 < s < 0.5
    w = exp(-U)
    W = (1-s) * exp(-tau *U)
    s = w/W - 1
    return 1 - (1-exp(-2*x*N*s))/(1-exp(-2*N*s))

def mean_fitness(t,x,s,U,tau):
    assert 0 <= t
    assert 0 <= x <= 1
    assert 0 < s < 0.5
    assert 0 < U < s
    assert 1 <= tau
    return x * exp(-U) + (1-x) * (1-s) * exp(-U - (tau - 1) * U * (1 - (1 - s)**t))

def phi(N,k,x):
    assert 1 <= N
    assert (0 <= k).all()
    assert 0 <= x <= 1
    if type(k) == ndarray:
        k[k>N] = N
    else:
        k = min(k,N)
    return stats.binom.pmf(k,N,x)
```
