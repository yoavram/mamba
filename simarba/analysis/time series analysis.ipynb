```python
import glob
import ujson as json
import os.path
import pandas as pd
import itertools
rcParams['figure.figsize'] = (8,6)
```

```python
output_dir = os.path.join('..','..','output','invasion')
param_files = glob.glob(os.path.join(output_dir, '*.json'))
param_files[:6]
```

```python
file_id = '2014-Jun-14_08-07-24-965130'
print "File ID:", file_id

# params
with open(os.path.join(output_dir, file_id + '.json')) as f:
    params = json.load(f)
assert params[u'sumatra_label'] == file_id
for k in ['in_pi','in_tau','in_phi','in_rho','r','pop_size']:
    print k, params[k]

# data
df = pd.read_csv(os.path.join(output_dir, file_id + '.csv.gz'), compression='gzip')
print 'Data shape:', df.shape

# log
with open(os.path.join(output_dir, file_id + '.log')) as f:
    log = f.readlines()
changes = []
for line in log:
    if 'Environmental change at tick' in line:
        i = line[::-1].index(' ')
        changes.append(int(line[-i:].strip()))
print 'Environmental changes at ticks', changes
```

```python
group_by_tick = df.groupby(by=('tick'))
mean_fitness_by_tick = group_by_tick.fitness.mean()
mean_fitness_by_tick.plot(color='k', ls='', marker='o')

def pretty_plot():
    axvline(x=params['in_tick'], color='k', ls='--')
    [axvline(x=x, color='gray', ls='--') for x in changes]
    fitness_lines = [(1-params['s'])**(params['envch_str']*x) for x in range(len(changes)+1)]
    [axhline(y=x, color='gray', ls='--') for x in fitness_lines]
    axhline(y=params['in_rate'], color='k', ls='--')
    ylim(-0.1,1.1)
    xlim(-params['final_tick']*.01,params['final_tick']*1.01)

pretty_plot()
ylabel('mean fitness');
```
